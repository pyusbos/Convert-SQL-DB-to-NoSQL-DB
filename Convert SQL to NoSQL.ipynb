# 2. Convert SQL database to NoSQL database

# 2.1 Design thought

Since I want to create a NoSQL database, I have to store the data in form of documments. Before digging depper into how to create them, we firstly understand two important concepts:

* Collections: Collections in NoSQL are analogus to the Tables in the SQL databases. Insert, Update and Delete operations can be performed on a collection.

* Documents: Each record of a master table in Relational database can be considered as a document here. For example in this restaurant database we have data for each restaurant and we store all the other tables in relation with the keys. In the NoSQL database, each restaurant will be a stand alone document on its own.

So, to convert SQL database to NoSQL database:

* step 1: Combine relational tables using primary keys.
* step 2: Design the structure of a restaurant document.
* step 3: Store all information of a restaurant in a document.
* step 4: Store all documents in a collection.
* step 5: Create a restaurant collection in MangoDB.
* step 6: Import all data into MangoDB.

Design the structure of a restaurant document:<br />
All information about a restaurant can be devided into three part. First part is the basic information about the restaurant, such as restaurant name and restaurant phone number. Second part is the address of the restaurant. An address includes three part: street address, city name and zipcode. Last part is the review information about the restaurant, such as review content and review number. This part includes social media information.

# 2.2 Preprocess data

I'm going to use loops to store data in dictionaries. To save running time, I firstly combine some tables using their keys.

```python
# Comine table "restaurant_category" and table "category"

temp_df = pd.merge(restaurant_category_df,category_df,on = "category_id")
category_df = temp_df
category_df.head()
```

```python
# Combine table "address" and table "city"

address_df = pd.DataFrame(restaurant_df[["restaurant_id","address","zipcode"]])
temp_df = pd.merge(address_df,city_df,on="zipcode")
address_df = temp_df
address_df.head()
```

```python
# Combine table "restaurant" and "address"

del restaurant_df["address"]
del restaurant_df["zipcode"]
temp_df = pd.merge(restaurant_df,address_df,on="restaurant_id")
restaurant_df = temp_df
restaurant_df.head()
```

```python
# Combine table "review" and "user"

temp_df = pd.merge(review_df,user_df,on="user_id")
review_df = temp_df
review_df.head()
```

# 2.3 Create documents

* The below code is used for converting our tables into documents which will further be stored in a collection. We use nested for loops to store all the data related to a particular movie in a single document. 

* The below structure could be understood as a list of nested dictionaries to store data related to a restaurant.

```python
# Define empty dictionaries and list which will hold the data inside.
restaurant_doc = []
restaurant_dict = {}

category = []

address_dict = {}

review = []
```

```python
# Create a document for a restaurant
for i in range(len(restaurant_df)):
    restaurant_dict = {"restaurant_id":restaurant_df['restaurant_id'][i],
                      "restaurant_name":restaurant_df['restaurant_name'][i],
                      "restaurant_url":restaurant_df['restaurant_url'][i],
                      "price":restaurant_df['price'][i],
                      "rating":float(restaurant_df['rating'][i]),
                      "phone":restaurant_df['phone'][i],
                      "review_count":int(restaurant_df["review_count"][i])}
    
    # Addd information of "category"
    for j in range(len(category_df)):
        if category_df["restaurant_id"][j] == restaurant_df["restaurant_id"][i]:
            cate = {"category":category_df["category_name"][j]}
            category.append(cate)
    
    category_list = {"categories":category}
    restaurant_dict.update(category_list)
    category = []
    
    # Add information of "address"
    address_dict = {"street_address":restaurant_df["address"][i],
                   "city":restaurant_df["city"][i],
                   "zipcode":restaurant_df["zipcode"][i]}
    
    address_list = {"address":[address_dict]}
    restaurant_dict.update(address_list)
    
    # Add information of "review"
    flag = 0
    for k in range(len(review_df)):
        if review_df["restaurant_id"][k] == restaurant_df["restaurant_id"][i]:
            rev = {"review_id":review_df["review_id"][k],
                  "review_rating":float(review_df["review_rating"][k]),
                  "review_text":review_df["text"][k],
                  "review_time":review_df["time_created"][k],
                  "user_id":review_df["user_id"][k],
                  "user_name":review_df["user_name"][k],
                  "user_profile":review_df["profile_url"][k]}
            review.append(rev)
            flag = flag+1
        
        if flag == 3:
            break
    
    review_list = {"reviews":review}
    restaurant_dict.update(review_list)
    review = []
      
    restaurant_doc.append(restaurant_dict)
```
